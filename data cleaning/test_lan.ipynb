```python
import re
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
file = open("loss.txt")
cnn_los = pd.DataFrame(columns=('step','loss'))
while 1:
    line = file.readline().replace(" ","")
    loss = re.findall(r"=(.+),",line)
    step = re.findall(r"p=(.+)\(",line)
    if loss and step:
        cnn_los= cnn_los.append({"step":int(step[0]),"loss":float(loss[0])},ignore_index=True)
    if not line:
        break
```

```python
plt.figure(figsize=(30,15))#
# plt.title('losss',fontsize=20)#
plt.xlabel('step',fontsize=24)#x
plt.ylabel('loss',fontsize=24)#y
plt.plot(cnn_los['step'],cnn_los['loss'])
```
